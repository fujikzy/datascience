```python
#求職者数
jobseeker_num = len(c_dt['jobseeker_no']) 

#RA推薦OKだった人数
ra_ok_num=c_dt[ c_dt['ra_recommend_flag'] == 1]['jobseeker_no'].count()

#書類選考OKだった人数
doc_ok_num=c_dt[ c_dt['doc_ok_flag'] == 1]['jobseeker_no'].count()

#ランダムでRA推薦OKを当てる確率
rand_ra_p = float(ra_ok_num) / float(jobseeker_num)

#ランダムで書類選考OKを当てる確率
rand_doc_p = float(doc_ok_num) / float(jobseeker_num)
```

```python
doc_ok_num
```

```python
#レコメンド順リスト
recommend_order = range(1, jobseeker_num+1) 

#RA推薦数リスト（ランダムの場合）
rand_ra = []
for i in range(1, jobseeker_num+1):
    rand_ra.append(rand_ra_p * i)

#書類選考OK数リスト（ランダムの場合）
rand_doc = []
for i in range(1, jobseeker_num+1):
    rand_doc.append(rand_doc_p * i)

#RA推薦数リスト（ルールベース）

print(rand_doc)

#    rand_ra.append(rand)
#rand_ra = range(rand_ra_p, ra_ok_num, rand_ra_p)
#rand_ra
```

##基礎集計

```python
#データ読み込み
c_dt= pd.read_csv('c_data.csv')
c_dt.head(10)
```

```python
#欠損確認1
len(c_dt) - c_dt.count()
```

```python
#欠損確認2
c_dt[c_dt.isnull().any(axis=1)]
```

```python
#欠損削除
c_dt_nan_d = c_dt.dropna()
len(c_dt_nan_d) - c_dt_nan_d.count()
```

```python
#基礎統計量
c_dt_nan_d.describe()
```

```python
#相関係数
c_dt_nan_d.corr()
```

```python
#標準化
c_dt_nan_d_n = c_dt_nan_d
c_dt_nan_d_n['age'] = stats.zscore(c_dt_nan_d['age'], axis=0)
c_dt_nan_d_n['education'] = stats.zscore(c_dt_nan_d['education'], axis=0)
c_dt_nan_d_n['scale'] = stats.zscore(c_dt_nan_d['scale'], axis=0)
c_dt_nan_d_n['job_length'] = stats.zscore(c_dt_nan_d['job_length'], axis=0)

c_dt_nan_d_n.head(10)
c_dt_nan_d.head(10)
#m = c_dt_nan_d.mean(1)
#s = c_dt_nan_d.std(1)
#c_dt_nan_d_n = c_dt_nan_d.sub(m,axis=0).div(s,axis=0)
```
